```python
aliases = pd.read_csv("data/7227.protein.aliases.v12.0.txt", sep="\t")
aliases.head()
```

```python
keep_same_proteins = []
for i in range(len(more)):
    row = more.iloc[i]
    first_col = row[0]
    first_col_string = aliases[aliases["alias"]==first_col]["#string_protein_id"].drop_duplicates()
    
    print(i)
    if len(first_col_string)>0:
        j=1
        write=True
        while row[j]!=None and j!=26:
            current = row[j]
            current_string = aliases[aliases["alias"]==current]["#string_protein_id"].drop_duplicates()
            if len(current_string)==0:
                write=False
                break
            if current_string.item() != first_col_string.item():
                write=False
                break
            j+=1
        if write:
            keep_same_proteins.append(row["Protein IDs"])
```
